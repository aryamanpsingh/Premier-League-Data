```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.common.keys import Keys
import csv
```

```python
browser = webdriver.Chrome(executable_path='/Users/aryamanpratapsingh/Desktop/py/chromedriver', chrome_options=webdriver.ChromeOptions())
browser.get("https://fbref.com/en/comps/9/stats/Premier-League-Stats")
# Wait 20 seconds for page to load
timeout = 200
try:
    # Wait until the final element [Avatar link] is loaded.
    # Assumption: If Avatar link is loaded, the whole page would be relatively loaded because it is among
    # the last things to be loaded.
    WebDriverWait(browser, timeout).until(EC.visibility_of_element_located((By.XPATH, "//*[@id='header']/a/img")))
except TimeoutException:
    print("Timed out waiting for page to load")
    browser.quit()
    
team_name_element = browser.find_elements_by_xpath(".//*[@id='stats_teams']/tbody/tr/th")
team_name = [x.text for x in team_name_element]

players_used_element = browser.find_elements_by_xpath("//*[@id='stats_teams']/tbody/tr/td[1]")
players_used = [x.text for x in players_used_element]

games_played_element = browser.find_elements_by_xpath("//*[@id='stats_teams']/tbody/tr/td[2]")
games_played = [x.text for x in games_played_element]

gs_element = browser.find_elements_by_xpath("//*[@id='stats_teams']/tbody/tr/td[3]")
gs = [x.text for x in gs_element]

assists_element = browser.find_elements_by_xpath("//*[@id='stats_teams']/tbody/tr/td[4]")
assists = [x.text for x in assists_element]

pk_scored_element = browser.find_elements_by_xpath("//*[@id='stats_teams']/tbody/tr/td[5]")
pk_scored = [x.text for x in pk_scored_element]

pk_won_element = browser.find_elements_by_xpath("//*[@id='stats_teams']/tbody/tr/td[6]")
pk_won = [x.text for x in pk_won_element]

fouls_element = browser.find_elements_by_xpath("//*[@id='stats_teams']/tbody/tr/td[7]")
fouls = [x.text for x in fouls_element]

yellow_cards_element = browser.find_elements_by_xpath("//*[@id='stats_teams']/tbody/tr/td[8]")
yellow_cards = [x.text for x in yellow_cards_element]

red_cards_element = browser.find_elements_by_xpath("//*[@id='stats_teams']/tbody/tr/td[9]")
red_cards = [x.text for x in red_cards_element]

shots_target_element = browser.find_elements_by_xpath("//*[@id='stats_teams']/tbody/tr/td[10]")
shots_target = [x.text for x in shots_target_element]

goals_pg_element = browser.find_elements_by_xpath("//*[@id='stats_teams']/tbody/tr/td[11]")
goals_pg = [x.text for x in goals_pg_element]

goals_asst_pg_element = browser.find_elements_by_xpath("//*[@id='stats_teams']/tbody/tr/td[12]")
goals_asst_pg = [x.text for x in goals_asst_pg_element]

goals_no_pen_pg_element = browser.find_elements_by_xpath("//*[@id='stats_teams']/tbody/tr/td[13]")
goals_no_pen_pg = [x.text for x in goals_no_pen_pg_element]

goals_asst_no_pen_pg_element = browser.find_elements_by_xpath("//*[@id='stats_teams']/tbody/tr/td[14]")
goals_asst_no_pen_pg = [x.text for x in goals_asst_no_pen_pg_element]

shots_pg_element = browser.find_elements_by_xpath("//*[@id='stats_teams']/tbody/tr/td[15]")
shots_pg = [x.text for x in shots_pg_element]

fouls_pg_element = browser.find_elements_by_xpath("//*[@id='stats_teams']/tbody/tr/td[16]")
fouls_pg = [x.text for x in fouls_pg_element]

cards_pg_element = browser.find_elements_by_xpath("//*[@id='stats_teams']/tbody/tr/td[17]")
cards_pg = [x.text for x in cards_pg_element]


zip_stats = list(zip(team_name,players_used,games_played,gs,assists,pk_scored,pk_won,fouls,yellow_cards,red_cards,shots_target,goals_pg,goals_asst_pg,goals_no_pen_pg,goals_asst_no_pen_pg,shots_pg,fouls_pg))

browser.close()
```
